To use `binscatter`, just make a file called `binscatter.py` in the same directory as your code, and copy-and-paste the contents of binscatter.py from here:

https://raw.githubusercontent.com/esantorella/binscatter/master/binscatter/binscatter.py

into your `binscatter.py` file. Then you will be able to import it using `import binscatter` like we do here

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import binscatter
```

Here I'm generating a bunch of random data, but with the relationship that y is *sort of* equal to 2 * x, except that there's a bunch of "noise" (random numbers added to or subtracted from 2 * x) to make the relationship look less clear when using a normal scatterplot

```python
x = np.random.normal(0, 1, 1000)
y = 2 * x + np.random.normal(0, 3.5, 1000)
```

```python
plt.scatter(x, y)
```

Now, the one annoying thing you need to do before you can use `binscatter`:

`binscatter` doesn't allow the simple `plt.binscatter()` syntax, sadly, but you just need to add this additional line:

`current_axis = plt.gca()`

then you call `current_axis.binscatter(x, y)` in the same way you would call `plt.scatter(x, y)`!

```python
current_axis = plt.gca()
current_axis.binscatter(x, y);
```

In this case, "binning" the values and then plotting the average of each bin shows the relationship between the variables in a much cleaner way than the standard `plt.scatter()` plot above, imo. From how well the line fits the points, you can tell that there is most likely a nice linear relationship between x and y here.
