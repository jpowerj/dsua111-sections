```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_career_anya(pid):
    print(pid)
    response = requests.get(f"https://www.pro-football-reference.com/players/{pid[0]}/{pid}.htm")
    soup = BeautifulSoup(response.text, 'html.parser')
    height_span = soup.find("span", attrs={'itemprop':'height'})
    if height_span is not None:
        height = height_span.text
        height_elts = height.split("-")
        height_ft = int(height_elts[0])
        height_in = int(height_elts[1])
    else:
        height_ft = np.nan
        height_in = np.nan
    weight_span = soup.find("span", attrs={'itemprop':'weight'})
    if weight_span is not None:
        weight = int(weight_span.text.replace("lb",""))
    else:
        weight = np.nan
    pass_tbl = soup.find("table", attrs={'id':'passing'})
    #print(pass_tbl)
    if pass_tbl is not None:
        pass_df = pd.read_html(str(pass_tbl))[0]
        career_anya = float(pass_df.loc[pass_df["Year"] == "Career","ANY/A"])
    else:
        career_anya = np.nan
    return pd.Series([height_ft, height_in, weight, career_anya])
```
